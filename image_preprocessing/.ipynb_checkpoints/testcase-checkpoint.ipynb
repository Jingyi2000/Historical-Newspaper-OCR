# Image Processing

In this step, the scanned pages need to be standardized. After analyzing the quality of our newspaper scans, the processing steps are determined as follows:
__graying, binarization, cropping (contour detection), skew correction, scaling__

## Step 1: Image cropping
OpenCV (Open Source Computer Vision Library) is an open-source computer vision and machine learning software library. It offers a rich set of image processing capabilities, including image transformation, filtering, feature detection, image segmentation, and more. Observations of scanned newspaper documents reveal the following characteristics: 1. The quality of the newspaper scans overall is good. 2. Some of the newspaper pages scanned include unwanted content (for example, a small part of the right page is included in the scan of the left page). 3. Some pages exhibit slight skewing.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def auto_crop_image(image_path): # Define a function for automated image cropping
    image = cv2.imread(image_path) 
    # Convert to grayscale image
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Apply thresholding to convert the image to a binary black and white image
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # Use contour detection algorithm: Identify black boundaries to separate the required individual pages
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    # cv2.RETR_EXTERNAL indicates that only external contours are detected
    # cv2.CHAIN_APPROX_SIMPLE compresses the contour boundary points to save memory
    
    cropped_image = None # Initialize an empty variable for storing the cropped image later
    
    # Check if contours exist, if so, choose one contour for cropping
    if contours:
        # Select the contour with the largest area
        sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)
        x, y, w, h = cv2.boundingRect(sorted_contours[0])
        # Crop the image based on the contour
        cropped_image = image[y:y+h, x:x+w]

    return image, cropped_image # Return both the original and cropped images for preliminary comparison
```

```python
# Use the function and directly display both the original and cropped images
original_image, cropped_image = auto_crop_image('/Users/yu/Desktop/linshi/test/00000006.tif.original.jpg')

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display the original image
axes[0].imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')

# Display the cropped image, if no cropping was done, display a text message instead
if cropped_image is not None:
    axes[1].imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    axes[1].set_title("Cropped Image")
    save_path = '/Users/yu/Desktop/linshi/result/00000006_cropped.jpg'
    cv2.imwrite(save_path, cropped_image)
else:
    axes[1].text(0.5, 0.5, 'No Change', horizontalalignment='center', verticalalignment='center')
axes[1].axis('off')

plt.show()
```

## Step 2: Skew correction
The skew of an image can potentially have a negative impact on subsequent OCR recognition, so correction is necessary. Since the skew in scanned documents is not severe, the correction logic employed first identifies the outer contours. If the contours exhibit skew, appropriate adjustments are made based on the degree of skew.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def correct_skew(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Detect black borders using edge detection
    _, thresh = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Calculate the skew angle
    if contours is not None and len(contours) > 0:
        largest_contour = max(contours, key=cv2.contourArea)
        rect = cv2.minAreaRect(largest_contour)
        angle = rect[2]
    else:
        angle = 0

    # Rotate the image to correct skew
    if abs(angle) > 0.1:
        if angle < 0:
            angle += 90
        else:
            angle -= 90
        rotation_matrix = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), angle, 1)
        image_corrected = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]), flags=cv2.INTER_LINEAR)
    else:
        image_corrected = image
    
    return image_corrected

# Call the function
input_image_path = '/Users/yu/Desktop/linshi/result/00000007_cropped.jpg'
corrected_image = correct_skew(input_image_path)

# Display the before-and-after comparison of skew correction
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(cv2.cvtColor(cv2.imread(input_image_path), cv2.COLOR_BGR2RGB))
axes[0].set_title("Cropped Image")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(corrected_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Corrected Image")
axes[1].axis('off')

plt.show()
```
